```
c=0
with open('train.txt','w') as train:
  for _,_,files in os.walk('.'):
    for file in files:
      if file.endswith('.txt') and file[:-4]+'.jpg' in files:
        train.write('/content/drive/My Drive/ANPR/darknet/data/numberPlateData/'+file[:-4]+'.jpg\n')
        c+=1
```

```
with open('test.txt','w') as test:
  for _,_,files in os.walk('.'):
    for file in files:
      if file.endswith('.jpeg'):
        test.write('/content/drive/My Drive/ANPR/darknet/data/numberPlateData'+'/'+file+'\n')
        c+=1
```
